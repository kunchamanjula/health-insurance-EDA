# 1.create different categorises based on their bmi values by defining new function?

```python
#categorzing the bmi values based on  weights
def bmi_categories(value):
    
    if value < 18.5:
        return "underweight"
    elif value >= 18.5 and value < 25:
        return "normal"
    elif value >= 25 and value <= 30:
        return "overweight"
    elif value > 30:
        return "obesity"
    else:
        return value

# now create the categories according to the values
insurance["bmi_cat"] = insurance["bmi"].apply(lambda x: bmi_categories(x)).copy( )
```

```python
insurance.head() # new column has been added to dataset on the name of bmi_cat
```

# 3. plot the barchart to  withrespect to index and differeint varaibles in dataset?

# 5.create different categorises based on their blood preesures  values by defining new function by the given data set? 

```python
#categorzing the blood pressures based on bp values
def bp_catogories(value):
     if value < 120:
        return "normal"
    
     elif value >= 120 and value < 130:
        return "elevated"
     elif value >= 130 and value < 140:
        return "hbp-stage1" # high blood pressure stage 1
     elif value >= 140 and value < 180:
        return "hbp-stage2" # high blood pressure stage 2
     elif value >= 180:
        return "hypertensive_crisis"
     else:
        return value


# now create the categories according to the values
insurance["bp_Cat"] = insurance["bloodpressure"].apply(lambda x: bp_catogories(x)).copy( )
```

```python
insurance.head(5)
```

```python
insurance.groupby(by="bp_Cat")["claim"].agg(func=["mean", "median", "max", "min"]).sort_values(by="mean", ascending=False)#using groupby aggratation to know the values of min max, median in blood preesure catorgies
```
